```python
X = df[["Temperature","Humidity","Rainfall","Wind_Speed","Soil_Moisture","Crop_Type","Water_Flow","Crop_Growth_Stage"]]
```

```python
X
```

```python
y = df["Crop_Health_Label"]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X["Crop_Type"] = le.fit_transform(X["Crop_Type"])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split (X, y,random_state = 42, test_size = 0.2)
```

```python
# Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
```

```python
lr_pred = lr.predict(X_test)
print("Logistic Regression accuracy_score", accuracy_score(y_test, lr_pred))
```

```python
# Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train,y_train)
```

```python
rf_pred = rf.predict(X_test)
print("Random Forest Accuracy", accuracy_score(y_test,rf_pred))
```

```python
# Naive bayes
```

```python
#1.GaGaussianNB
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
gnb_pred = gnb.predict(X_test)
print("GaussianNB Accuracy", accuracy_score(y_test, gnb_pred))
```

```python
#2.MultinomialNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
```

```python
# scaler = MinMaxScaler()
# X_train_m = scaler.fit_transform(X_train)
# X_test_m = scaler.transform(X_test)
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
```

```python
mnb_pred = mnb.predict(X_test)
print("Multinomial Accuracy", accuracy_score(y_test,mnb_pred))
```

```python
#3.BernoulliNB
from sklearn.naive_bayes import BernoulliNB
```

```python
bnb = BernoulliNB()
bnb.fit(X_test,y_test)
```

```python
bnb_pred = bnb.predict(X_test)
print("BernoulliNB Accuracy", accuracy_score(y_test, bnb_pred))
```

```python
#Model Comparison
```

```python
print("MODEL COMPARISON")
print("----------------------------")
print("Logistic Regression :", accuracy_score(y_test, lr_pred))
print("Random Forest       :", accuracy_score(y_test, rf_pred))
print("Gaussian Naive Bayes:", accuracy_score(y_test, gnb_pred))
print("Multinomial NB      :", accuracy_score(y_test, mnb_pred))
print("Bernoulli NB        :", accuracy_score(y_test, bnb_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

# Model names
models = [
    "Logistic Regression",
    "Random Forest",
    "Gaussian NB",
    "Multinomial NB",
    "Bernoulli NB"
]

# Accuracy values
accuracies = [
    accuracy_score(y_test, lr_pred),
    accuracy_score(y_test, rf_pred),
    accuracy_score(y_test, gnb_pred),
    accuracy_score(y_test, mnb_pred),
    accuracy_score(y_test, bnb_pred)
]

# Plot bar graph
plt.figure(figsize=(8,5))
plt.bar(models, accuracies)
plt.xlabel("Machine Learning Models")
plt.ylabel("Accuracy")
plt.title("Model Comparison for Crop Health Prediction")
plt.xticks(rotation=30)
plt.ylim(0,1)
plt.show()
```
